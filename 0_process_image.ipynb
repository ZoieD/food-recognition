```python
import glob
import os
import shutil
```

```python
images = glob.glob("./data/custom/images/train/*")
```

```python
images
```

```python
for i in images:
    pre_name = i.split("/")
    base_dir = i.split(pre_name[-1])
    new_name = pre_name[-2] + '_' + pre_name[-1]
    os.rename(i, base_dir[0] + new_name)
    shutil.move(base_dir[0] + new_name, i.split(pre_name[-2])[0] + '/' + new_name)
```
